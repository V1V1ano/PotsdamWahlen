```python
import pandas as pd
```

```python
df = pd.read_csv('Tables/lwerst2019.csv', sep=';', usecols=['Nr', 'Wahlkreis', 'Name', 'Wahlb. insges.', 'Wähler'])
df.head()
```

```python
#Filterraus Briefwahlbezirk

substring = 'Briefwahlbezirk'
filter = df['Name'].str.contains(substring)
filtered_df = df[~filter]
filtered_df
```

```python
# Add Wahlbeteiligung

filtered_df['Wahlbeteiligung'] = filtered_df[['Wähler']].div(filtered_df['Wahlb. insges.'], axis=0)
filtered_df['Wahlbeteiligung'] = filtered_df['Wahlbeteiligung'] *100
filtered_df['Wahlbeteiligung'] = filtered_df['Wahlbeteiligung'].astype(float).round(2)


filtered_df.head()
```

```python
# repeat for Kommunalwahlen 2024

df1 = pd.read_csv('Tables/kw2024.csv', sep=';', usecols=['Gebietsart', 'Gebietsname', 'Bezirksnummer', 'Wahlberechtigte gesamt (A)', 'Waehler gesamt (B)'])
filter = df1['Gebietsname'].str.contains(substring)
filtered_df1 = df1[~filter]

#Wahlbeteiligung hinzufuegen
filtered_df1['Wahlbeteiligung_kw2024'] = filtered_df1[['Waehler gesamt (B)']].div(filtered_df1['Wahlberechtigte gesamt (A)'], axis=0)
filtered_df1['Wahlbeteiligung_kw2024'] = filtered_df1['Wahlbeteiligung_kw2024'] *100
filtered_df1['Wahlbeteiligung_kw2024'] = filtered_df1['Wahlbeteiligung_kw2024'].astype(float).round(2)

filtered_df1.head()
```

```python
# Prepare Tables for merge
# rename to distinct

lwerst2019_df = filtered_df
kw2024_df = filtered_df1

# drop unnecessary columns
kw2024_df = kw2024_df.drop(columns=['Gebietsart', 'Gebietsname', 'Wahlberechtigte gesamt (A)', 'Waehler gesamt (B)'])
lwerst2019_df = lwerst2019_df.drop(columns=['Wahlkreis', 'Wahlb. insges.', 'Wähler'])
lwerst2019_df = lwerst2019_df.rename(columns={"Nr" : "Bezirksnummer", "Wahlbeteiligung" : "Wahlbeteiligung_lwerst219"})

combined_df = pd.merge(lwerst2019_df, kw2024_df, on="Bezirksnummer")
combined_df
```
