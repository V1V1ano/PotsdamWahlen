```python
import pandas as pd
```

```python
df = pd.read_csv('Tables/lwerst2019.csv', sep=';', usecols=['Nr', 'Wahlkreis', 'Name', 'Wahlb. insges.', 'Wähler'])
df.head()
```

```python
#Filterraus Briefwahlbezirk

substring = 'Briefwahlbezirk'
filter = df['Name'].str.contains(substring)
filtered_df = df[~filter]
filtered_df
```

```python
# Add Wahlbeteiligung

filtered_df['Wahlbeteiligung'] = filtered_df[['Wähler']].div(filtered_df['Wahlb. insges.'], axis=0)
filtered_df['Wahlbeteiligung'] = filtered_df['Wahlbeteiligung'] *100
filtered_df['Wahlbeteiligung'] = filtered_df['Wahlbeteiligung'].astype(float).round(2)


filtered_df.head()
```

```python
# repeat for Kommunalwahlen 2024

df1 = pd.read_csv('Tables/kw2024.csv', sep=';', usecols=['Gebietsart', 'Gebietsname', 'Gebietsnummer', 'Bezirksnummer', 'Wahlberechtigte gesamt (A)', 'Waehler gesamt (B)'])
filter = df1['Gebietsname'].str.contains(substring)
filtered_df1 = df1[~filter]

#Wahlbeteiligung hinzufuegen
filtered_df1['Wahlbeteiligung_kw2024'] = filtered_df1[['Waehler gesamt (B)']].div(filtered_df1['Wahlberechtigte gesamt (A)'], axis=0)
filtered_df1['Wahlbeteiligung_kw2024'] = filtered_df1['Wahlbeteiligung_kw2024'] *100
filtered_df1['Wahlbeteiligung_kw2024'] = filtered_df1['Wahlbeteiligung_kw2024'].astype(float).round(2)

filtered_df1.head()
```

```python
# repeat for Europawahlen 2024

df2 = pd.read_csv('Tables/euro2024.csv', sep=';', usecols=['Gebietsart', 'Gebietsname', 'Bezirksnummer', 'Wahlberechtigte gesamt (A)', 'Waehler gesamt (B)'])
filter = df2['Gebietsart'].str.contains("BRIEFWAHLBEZIRK")
filtered_df2 = df2[~filter]

#Wahlbeteiligung hinzufuegen
filtered_df2['Wahlbeteiligung_euro2024'] = filtered_df2[['Waehler gesamt (B)']].div(filtered_df2['Wahlberechtigte gesamt (A)'], axis=0)
filtered_df2['Wahlbeteiligung_euro2024'] = filtered_df2['Wahlbeteiligung_euro2024'] *100
filtered_df2['Wahlbeteiligung_euro2024'] = filtered_df2['Wahlbeteiligung_euro2024'].astype(float).round(2)

filtered_df2
```

```python
# repeat for Europwahl 2019
df3 = pd.read_csv('Tables/ew2019.csv', sep=';', usecols=['Nr', 'Name', 'Wahlb. insges.', 'Wähler'])

# filter out Briefwahlbezirk

substring = 'Briefwahlbezirk'
filter = df3['Name'].str.contains(substring)
filtered_df3 = df3[~filter]

# Add Wahlbeteiligung

filtered_df3['Wahlbeteiligung_euro2019'] = filtered_df3[['Wähler']].div(filtered_df3['Wahlb. insges.'], axis=0)
filtered_df3['Wahlbeteiligung_euro2019'] = filtered_df3['Wahlbeteiligung_euro2019'] *100
filtered_df3['Wahlbeteiligung_euro2019'] = filtered_df3['Wahlbeteiligung_euro2019'].astype(float).round(2)

```

```python
# repeat for Kommunalwahl 2019
df4 = pd.read_csv('Tables/kw2019.csv', sep=';', usecols=['Nr', 'Name', 'Wahlb. insges.', 'Wähler'])

# filter out Briefwahlbezirk

substring = 'Briefwahlbezirk'
filter = df4['Name'].str.contains(substring)
filtered_df4 = df4[~filter]

# Add Wahlbeteiligung

filtered_df4['Wahlbeteiligung_kw2019'] = filtered_df4[['Wähler']].div(filtered_df4['Wahlb. insges.'], axis=0)
filtered_df4['Wahlbeteiligung_kw2019'] = filtered_df4['Wahlbeteiligung_kw2019'] *100
filtered_df4['Wahlbeteiligung_kw2019'] = filtered_df4['Wahlbeteiligung_kw2019'].astype(float).round(2)
filtered_df4
```

```python
# repeat for Bundestagswahl 2021

df5 = pd.read_csv('Tables/bw2021.csv', sep=';', usecols=['Gebietsnummer', 'Wahlberechtigte gesamt (A)', 'Gebietsart', 'Waehler gesamt (B)'])

# filter out Briefwahlbezirk

substring = 'BRIEFWAHLBEZIRK'
filter = df5['Gebietsart'].str.contains(substring)
filtered_df5 = df5[~filter]

# Add Wahlbeteiligung

filtered_df5['Wahlbeteiligung_bw2021'] = filtered_df5[['Waehler gesamt (B)']].div(filtered_df5['Wahlberechtigte gesamt (A)'], axis=0)
filtered_df5['Wahlbeteiligung_bw2021'] = filtered_df5['Wahlbeteiligung_bw2021'] *100
filtered_df5['Wahlbeteiligung_bw2021'] = filtered_df5['Wahlbeteiligung_bw2021'].astype(float).round(2)
filtered_df5
```

```python
# Prepare Tables for merge
# rename to distinct

lwerst2019_df = filtered_df
kw2024_df = filtered_df1
euro2024_df = filtered_df2
euro2019_df = filtered_df3
kw2019_df = filtered_df4
bw2021_df = filtered_df5

# drop unnecessary columns
kw2024_df = kw2024_df.drop(columns=['Gebietsart', 'Gebietsname', 'Wahlberechtigte gesamt (A)', 'Waehler gesamt (B)'])
lwerst2019_df = lwerst2019_df.drop(columns=['Wahlkreis', 'Wahlb. insges.', 'Wähler'])
lwerst2019_df = lwerst2019_df.rename(columns={"Nr" : "Bezirksnummer", "Wahlbeteiligung" : "Wahlbeteiligung_lwerst2019"})
euro2024_df = euro2024_df.drop(columns=['Gebietsart', 'Gebietsname', 'Wahlberechtigte gesamt (A)', 'Waehler gesamt (B)'])
euro2019_df = euro2019_df.drop(columns=['Name', 'Wahlb. insges.', 'Wähler'])
euro2019_df = euro2019_df.rename(columns={"Nr" : "Bezirksnummer"})
kw2019_df = kw2019_df.rename(columns={"Nr" : "Bezirksnummer"})
kw2019_df = kw2019_df.drop(columns=['Name', 'Wahlb. insges.', 'Wähler'])
bw2021_df = bw2021_df.drop(columns=['Gebietsart', 'Wahlberechtigte gesamt (A)', 'Waehler gesamt (B)'])

#combined_df

combined_df = pd.merge(bw2021_df, kw2024_df, on="Gebietsnummer")
combined_df = combined_df.drop(columns=["Gebietsnummer"])
combined_df = pd.merge(combined_df, euro2019_df, on="Bezirksnummer")
combined_df = pd.merge(combined_df, euro2024_df, on="Bezirksnummer")
combined_df = pd.merge(combined_df, lwerst2019_df, on="Bezirksnummer")
combined_df = pd.merge(combined_df, kw2019_df, on="Bezirksnummer")

# euro2019, lw2019, kw2019, bw2021, kw2024, euro2024

combined_df = combined_df.loc[:,['Bezirksnummer','Name','Wahlbeteiligung_kw2019', 'Wahlbeteiligung_euro2019', 'Wahlbeteiligung_lwerst2019', 'Wahlbeteiligung_bw2021', 'Wahlbeteiligung_kw2024', 'Wahlbeteiligung_euro2024']]

```
