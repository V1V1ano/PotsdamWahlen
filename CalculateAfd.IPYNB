```python
import pandas as pd
```

```python
# Alternative für Deutschland Europwahl 2019
euro2019 = pd.read_csv('Tables/ew2019.csv', sep=';', usecols=['Nr', 'Name', 'Wahlb. insges.', 'Alternative für Deutschland'])

# Filter Briefwahlbezirk
filter = euro2019['Name'].str.contains('Briefwahlbezirk')
euro2019 = euro2019[~filter]

# add Wahlbeteiligung für Afd
euro2019['euro2019_afd'] = euro2019[['Alternative für Deutschland']].div(euro2019['Wahlb. insges.'], axis=0) * 100
euro2019['euro2019_afd'] = euro2019['euro2019_afd'].astype(float).round(2)

# drop unnecessary columns
euro2019 = euro2019.drop(columns=["Wahlb. insges.", "Alternative für Deutschland", "Name"])
```

```python
# Landtagswahl 2019
lw2019 = pd.read_csv("Tables/lwerst2019.csv", sep=';', usecols=["Name", "Wahlb. insges.", "Alternative für Deutschland", "Nr"])

# Filter Briefwahlbezirk
filter = lw2019['Name'].str.contains('Briefwahlbezirk')
lw2019 = lw2019[~filter]

# add Wahlbeteiligung für Afd
lw2019['lw2019_afd'] = lw2019[['Alternative für Deutschland']].div(lw2019['Wahlb. insges.'], axis=0) * 100
lw2019['lw2019_afd'] = lw2019['lw2019_afd'].astype(float).round(2)

# drop unnecessary columns
lw2019 = lw2019.drop(columns=["Wahlb. insges.", "Alternative für Deutschland"])
```

```python
# Kommunalwahl 2019
kw2019 = pd.read_csv("Tables/kw2019.csv", sep=';', usecols=["Name", "Wahlb. insges.", "Alternative für Deutschland", "Nr"])

# Filter Briefwahlbezirk
filter = kw2019['Name'].str.contains('Briefwahlbezirk')
kw2019 = kw2019[~filter]

# add Wahlbeteiligung für Afd
kw2019['kw2019_afd'] = kw2019[['Alternative für Deutschland']].div(kw2019['Wahlb. insges.'], axis=0) * 100
kw2019['kw2019_afd'] = kw2019['kw2019_afd'].astype(float).round(2)

# drop unnecessary columns
kw2019 = kw2019.drop(columns=["Wahlb. insges.", "Alternative für Deutschland", "Name"])


```

```python
# f2 ist AFD bei Bundestagswahl 2021
bw2021 = pd.read_csv('Tables/bw2021.csv', sep=';', usecols=['Gebietsnummer', 'Wahlberechtigte gesamt (A)', 'Gebietsart', 'F2'])

# Filter Briefwahlbezirk
filter = bw2021['Gebietsart'].str.contains('BRIEFWAHLBEZIRK')
bw2021 = bw2021[~filter]

# add Wahlbeteiligung für Afd
bw2021['bw2021_afd'] = bw2021[['F2']].div(bw2021['Wahlberechtigte gesamt (A)'], axis=0) * 100
bw2021['bw2021_afd'] = bw2021['bw2021_afd'].astype(float).round(2)

# drop unnecessary columns
bw2021 = bw2021.drop(columns=["Wahlberechtigte gesamt (A)", "Gebietsart", "F2"])

bw2021
```

```python
# D6 ist AFD bei Kommunalwahl 2024 (3 Stimmen koennen abgegeben werden)

kw2024 = pd.read_csv('Tables/kw2024.csv', sep=';', usecols=['Gebietsname', 'Gebietsnummer', 'Bezirksnummer', 'Wahlberechtigte gesamt (A)', 'D6'])

# Filter Briefwahl
filter = kw2024['Gebietsname'].str.contains('Briefwahlbezirk')
kw2024 = kw2024[~filter]

# add Wahlbeteiligung für Afd
kw2024['afd_durch3'] = kw2024[['D6']].div(3)
kw2024['kw2024_afd'] = kw2024[['afd_durch3']].div(kw2024['Wahlberechtigte gesamt (A)'], axis=0) * 100
kw2024['kw2024_afd'] = kw2024['kw2024_afd'].astype(float).round(2)

# drop unnecessary columns
kw2024 = kw2024.drop(columns=["Wahlberechtigte gesamt (A)", "Gebietsname", "D6", "afd_durch3"])

kw2024
```

```python
# Europawahl 2024, D1 = Afd

euro2024 = pd.read_csv('Tables/euro2024.csv', sep=';', usecols=['Gebietsart', 'Gebietsname', 'Bezirksnummer', 'Wahlberechtigte gesamt (A)', 'D1'])

# Filter Briefwahl
filter = euro2024['Gebietsart'].str.contains("BRIEFWAHLBEZIRK")
euro2024 = euro2024[~filter]

# add Wahlbeteiligung für Afd
euro2024['euro2024_afd'] = euro2024[['D1']].div(euro2024['Wahlberechtigte gesamt (A)'], axis=0) * 100
euro2024['euro2024_afd'] = euro2024['euro2024_afd'].astype(float).round(2)

# drop unnecessary columns
euro2024 = euro2024.drop(columns=["Wahlberechtigte gesamt (A)", "Gebietsname", "D1", "Gebietsart", "Gebietsname"])

euro2024
```

```python
# Merge

combined = pd.merge(bw2021, kw2024, on="Gebietsnummer")
combined = combined.drop(columns=["Gebietsnummer"])

euro2019 = euro2019.rename(columns={"Nr" : "Bezirksnummer"})
combined = pd.merge(combined, euro2019, on="Bezirksnummer")

combined = pd.merge(combined, euro2024, on="Bezirksnummer")

lw2019 = lw2019.rename(columns={"Nr" : "Bezirksnummer"})
combined = pd.merge(combined, lw2019, on="Bezirksnummer")

kw2019 = kw2019.rename(columns={"Nr" : "Bezirksnummer"})
combined = pd.merge(combined, kw2019, on="Bezirksnummer")

combined = combined.loc[:,['Bezirksnummer','Name','kw2019_afd', 'kw2019_afd', 'lw2019_afd', 'bw2021_afd', 'kw2024_afd', 'euro2024_afd']]

combined
```
